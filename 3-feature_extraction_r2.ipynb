```python
import pandas as pd
df=pd.read_csv("combined_comment.csv")
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
df=df.dropna()
```

```python

# drop all rows with float values in the 'Reviews' column
df = df[pd.to_numeric(df['Reviews'], errors='coerce').isna()]
# filter the dataframe to keep only sentences with more than 10 words
df_10 = df[df['Reviews'].apply(lambda x: len(x.split()) > 10)]
df_10
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# Remove any rows with missing values
df_10.dropna(subset=['Reviews'], inplace=True)
# Convert the comments into a bag-of-words representation
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df_10['Reviews'])

# Fit an LDA model to the data
lda = LatentDirichletAllocation(n_components=10, random_state=42)
X_topics = lda.fit_transform(X)
```

```python
# Print the top 10 words for each topic
feature_names = vectorizer.get_feature_names()
for topic_idx, topic in enumerate(lda.components_):
    print(f"Topic {topic_idx}:")
    top_words_idx = topic.argsort()[::-1][:10] # get the indices of the top 10 words for the topic
    top_words = [feature_names[i] for i in top_words_idx]
    print(top_words)
    print()
```

```python
# define the keywords for each topic as a list
topic_keywords = [
    ['like', 'look', 'car', 'wheel', 'thing', 'really', 'think', 'seat', 'make', 'good'],
    ['car', 'tesla', 'model', 'like', 'year', 'drive', 'new', 'love', 'ev', 'great'],
    ['dealer', 'nissan', 'dealership', 'car', 'like', 'msrp', 'order', 'said', 'video', 'new'],
    ['charger', 'charging', 'charge', 'battery', 'station', 'home', 'ev', 'need', 'use', 'power'],
    ['car', 'ev', 'people', 'year', 'make', 'brake', 'used', 'like', 'think', 'dont'],
    ['battery', 'mile', 'range', 'car', 'charge', 'leaf', 'time', 'trip', 'charging', 'driving'],
    ['tesla', 'ev', 'car', 'company', 'make', 'vehicle', 'people', 'market', 'like', 'year'],
    ['tire', 'car', 'ev', 'road', 'driving', 'vehicle', 'ice', 'use', 'need', 'speed'],
    ['model', 'price', 'range', 'awd', 'new', 'like', 'truck', 'ev', 'tax', 'ev6'],
    ['car', 'tax', 'credit', 'app', 'update', 'phone', 'time', 'got', 'dealer', 'week']
]

# loop over each topic and create a new dataframe containing only rows with key words in the topic
for i, keywords in enumerate(topic_keywords):
    topic_df = df_10[df_10['Reviews'].str.contains('|'.join(keywords))]
```

```python
# remove duplicates based on a subset of columns
topic_df= topic_df.drop_duplicates(subset=['Reviews'])
topic_df
```
