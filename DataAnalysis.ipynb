# 3. Language distribution in social good projects VS general projects

```python
data1=aggre[['Language','SocialGood']].dropna()
data2=aggre[['Language1','SocialGood']].dropna()
data2=data2.rename(columns={'Language1':'Language'})
data3=aggre[['Language2','SocialGood']].dropna()
data3=data3.rename(columns={'Language2':'Language'})
data4=aggre[['Language3','SocialGood']].dropna()
data4=data4.rename(columns={'Language3':'Language'})
plot=pd.concat([data1,data2,data3,data4],ignore_index=True)
```

```python
ploty=plot[plot['SocialGood']=='Y']
plotn=plot[plot['SocialGood']=='N']

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.pie(ploty['Language'].value_counts()[:10],labels=list(ploty['Language'].value_counts().to_frame().index)[0:10],autopct='%1.2f%%')
plt.title("Language Distribution in Social Good Projects")


plt.subplot(122)
plt.pie(plotn['Language'].value_counts()[:10],labels=list(plotn['Language'].value_counts().to_frame().index)[0:10],autopct='%1.2f%%') 
plt.title("Language Distribution in General Projects")

plt.show() 

# There is no significant difference
```

# 6. Will preference for project topics change over time


```python
time=aggre.groupby([aggre['Date'],aggre['Topic']])['document'].count().reset_index()
time["Date"] = time["Date"].map({"W18":"2018/04", "W19":"2019/04","F19":"2019/12","W20":"2020/04","F20":"2020/12","W21":"2021/04",
                                "F21":"2021/12","W22":"2022/04"})



time=time.sort_values(by="Date", ascending=True).reset_index()

total=time.groupby('Date')['document'].sum().reset_index()
time=time.merge(total, on='Date', how='left')
time['Proportion']=time['document_x']/time['document_y']

Development = time[time['Topic']=='Development/Living']
Economy = time[time['Topic']=='Economy/Finance']

Education = time[time['Topic']=='Education']
Entertainment = time[time['Topic']=='Entertainment']

Technology = time[time['Topic']=='Technology/Tool']
Health = time[time['Topic']=='Health']
```

```python
import matplotlib.pyplot as plt

plt.plot(Development['Date'],Development['Proportion'],'s-',color = 'r',label="Development")#s-:方形
plt.plot(Economy['Date'],Economy['Proportion'],'o-',color = 'g',label="Economy")#o-:圆形

plt.plot(Education['Date'],Education['Proportion'],'s-',color = 'black',label="Education")
plt.plot(Entertainment['Date'],Entertainment['Proportion'],'o-',color = 'blue',label="Entertainment")

plt.plot(Technology['Date'],Technology['Proportion'],'s-',color = 'y',label="Technology")
plt.plot(Health['Date'],Health['Proportion'],'o-',color = 'purple',label="Health")

plt.xlabel("Date")
plt.ylabel("Proportion")
plt.legend(loc = "best")
plt.show()
```
